```python
#Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
from scipy.stats import linregress
```

```python
#Air quality data
aq2020_df = pd.read_csv("Resources/aq2020.csv")
aq2019Q1_df = pd.read_csv("Resources/aq2019Q1.csv")
aq2019Q2_df = pd.read_csv("Resources/aq2019Q2.csv")
aq2019Q3_df = pd.read_csv("Resources/aq2019Q3.csv")
aq2019_df = pd.concat([aq2019Q1_df, aq2019Q2_df, aq2019Q3_df], join='outer')

#Format dataframe dates
date_format = [aq2020_df,aq2019_df]
for df in date_format:
    df['Date'] = pd.to_datetime(df['Date'])

aq2019_df = aq2019_df.dropna()
aq2020_df = aq2020_df.dropna()
```

```python
#Covid API URL
base_url = "https://api.covid19api.com/dayone/country/"
end_url = "/status/confirmed"

#Enter country
country = "USA"

response = requests.get(base_url+country+end_url).json()
response
#On responses, run following code:
USA_raw = pd.DataFrame.from_dict(response, orient='columns')

#Format covid dataframe dates (now mergable with air quality dataframes)
USA_raw['Date'] = USA_raw['Date'].str.strip('T00:00:00Z')
USA_raw['Date'] = pd.to_datetime(USA_raw['Date'])
USA_raw
```

```python
covid_boston = USA_raw.loc[(USA_raw['Province']=='Massachusetts') & (USA_raw['City']=='Suffolk')]
covid_boston = covid_boston.drop_duplicates(["City", "Date"], keep = 'first')
covid_boston = covid_boston[['Date', 'Cases']].groupby('Date').sum()

covid_char = USA_raw.loc[(USA_raw['Province']=='North Carolina') & (USA_raw['City']=='Mecklenburg')]
covid_char = covid_char.drop_duplicates(["City", "Date"], keep = 'first')
covid_char = covid_char[['Date', 'Cases']].groupby('Date').sum()

covid_la = USA_raw.loc[(USA_raw['Province']=='California') & (USA_raw['City']=='Los Angeles')]
covid_la = covid_la.drop_duplicates(["City", "Date"], keep = 'first')
covid_la = covid_la[['Date', 'Cases']].groupby('Date').sum()

covid_seattle = USA_raw.loc[(USA_raw['Province']=='Washington') & (USA_raw['City']=='King')]
covid_seattle = covid_seattle.drop_duplicates(["City", "Date"], keep = 'first')
covid_seattle = covid_seattle[['Date', 'Cases']].groupby('Date').sum()

covid_chicago = USA_raw.loc[(USA_raw['Province']=='Illinois') & (USA_raw['City']=='Cook') | (USA_raw['City']=='DuPage')]
covid_chicago = covid_chicago.drop_duplicates(["City", "Date"], keep = 'first')
covid_chicago = covid_chicago[['Date', 'Cases']].groupby('Date').sum()

covid_omaha = USA_raw.loc[(USA_raw['Province']=='Nebraska') & (USA_raw['City']=='Douglas')]
covid_omaha = covid_omaha.drop_duplicates(["City", "Date"], keep = 'first')
covid_omaha = covid_omaha[['Date', 'Cases']].groupby('Date').sum()
```
